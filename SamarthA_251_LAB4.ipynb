```python
'''Write a Python program to implement the object-oriented concepts of multiple, Multilevel and Hierarchical Inheritances using your domain applications.'''
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


class Patient(Person):
    def __init__(self, name, age, patient_id):
        super().__init__(name, age)
        self.patient_id = patient_id

    def display(self):
        super().display()
        print("Patient ID:", self.patient_id)


class Doctor(Person):
    def __init__(self, name, age, doctor_id, specialty):
        super().__init__(name, age)
        self.doctor_id = doctor_id
        self.specialty = specialty

    def display(self):
        super().display()
        print("Doctor ID:", self.doctor_id)
        print("Specialty:", self.specialty)


class Surgeon(Doctor):
    def __init__(self, name, age, doctor_id, specialty, procedures):
        super().__init__(name, age, doctor_id, specialty)
        self.procedures = procedures

    def display(self):
        super().display()
        print("Procedures:", ', '.join(self.procedures))


class Nurse(Person):
    def __init__(self, name, age, nurse_id):
        super().__init__(name, age)
        self.nurse_id = nurse_id

    def display(self):
        super().display()
        print("Nurse ID:", self.nurse_id)


# Usage
patient = Patient("John Doe", 30, "P123")
doctor = Doctor("Dr. Smith", 45, "D567", "Cardiologist")
surgeon = Surgeon("Dr. Johnson", 50, "D789", "Surgeon", ["Appendectomy", "Laparoscopy"])
nurse = Nurse("Jane Smith", 28, "N456")

patient.display()
print("\n")
doctor.display()
print("\n")
surgeon.display()
print("\n")
nurse.display()
```
